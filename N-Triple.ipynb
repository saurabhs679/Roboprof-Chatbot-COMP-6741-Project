```python
import rdflib
from rdflib import Graph

# Create a new RDF graph
g1 = Graph()

# Parse the TTL file and merge it into the graph
g1.parse("merged_data.ttl")
```

```python
nt_data = g1.serialize(format='nt')

# Print or save the N-Triples data as required
print(nt_data)
```

```python
import rdflib


nt_data = g1.serialize(format='nt')

file_path = 'merged_data.nt'  

with open(file_path, 'w') as file:
    file.write(nt_data)

print(f"Saved N-Triples data to {file_path}")
```
