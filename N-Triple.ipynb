```python
from rdflib import Graph, Namespace
from rdflib.plugins.sparql import prepareQuery

# Load the RDF data
g = Graph()
g.parse("output_new.ttl", format="ttl")

# Define the namespace prefix
ex = Namespace("http://example.org/")

# Define the SPARQL query
query = prepareQuery("""
    PREFIX ex: <http://example.org/>
    SELECT (COUNT(DISTINCT ?topic) as ?totalTopics)
    WHERE {
      ?topic a ex:Topic .
    }
""", initNs={"ex": ex})

# Execute the query
results = g.query(query)

# Extract the total number of distinct topics
for row in results:
    total_topics = row['totalTopics']

print("Total number of distinct topics:", total_topics)
```

```python
query = prepareQuery("""
    PREFIX ex: <http://example.org/>
    SELECT ?topic_in_course (COUNT(?topic) as ?topicInstances)
    WHERE {
      ?topic a ex:Topic ;
             ex:topic_in_course ?topic_in_course .
    }
    GROUP BY ?topic_in_course
""", initNs={"ex": ex})

# Execute the query
results = g.query(query)

# Print the number of topic instances per topic in course
for row in results:
    topic_in_course = row['topic_in_course']
    topic_instances = row['topicInstances']
    print(f"Topic in Course: {topic_in_course}, Number of Topic Instances: {topic_instances}")
```
